```python
#impoting libraries
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
import pandas as pd
import warnings
import matplotlib.cbook
import seaborn as sns
import numpy as np
import scipy.stats as stats
```

```python
#load data
crime_df=pd.read_csv("crimeratesbystate-formatted.csv")
```

#### 1.Histogram

```python
crime_df.hist(column='larceny_theft',bins=10,color='blue',alpha=0.8,label='Value',edgecolor='green',linewidth=1.5)
plt.xlabel("Larceny Theft")
plt.ylabel("Theft Counts")
plt.title("Histogram", size=10)
plt.savefig("Histogram Python.png")
plt.show()
```

#### 2. Box Plot

```python
fig, ax = plt.subplots()
boxplot = crime_df.boxplot(column=['larceny_theft'], return_type='axes')
labels = [item.get_text() for item in ax.get_xticklabels()]
labels[0] = 'Larceny Theft'
labels[1] = 'Burglary'
labels[2] = 'Aggravated Assault'
ax.set_xticklabels(labels)
plt.xlabel("Crime Theft")
plt.ylabel("Crime Counts")
plt.title("Box Plot", size=10)
plt.savefig("Box Plot Python.png")
plt.show()
```

#### 3. bullet graph

```python
edu_df=pd.read_csv("education.csv")
```

```python
eduudf = edu_df.sort_values(['math'], ascending=False).head(5)
eduudf = eduudf.reset_index()
edudf = eduudf[['state','math']]
edoutdf = edudf.copy()
edoutdf["tgtval"] = 800
edlst = edoutdf.values.tolist()
```

```python
limits=[500,600,700,800]
labels=["Poor", "OK", "Good", "Excellent"]
size=(8,5)
axis_label="Performance Measure"
label_color="black"
bar_color="#252525"
target_color='#f7f7f7'
title="Bullet graph"
palette=None
formatter=None
```

```python
h = limits[-1] / 10

if palette is None:
    palette = sns.light_palette("green", len(limits), reverse=False)

if len(edlst) == 1:
    fig, ax = plt.subplots(figsize=size, sharex=True)
else:
    fig, axarr = plt.subplots(len(edlst), figsize=size, sharex=True)

for idx, item in enumerate(edlst):
    if len(edlst) > 1:
        ax = axarr[idx]

    ax.set_aspect('equal')
    ax.set_yticklabels([item[0]])
    ax.set_yticks([1])
    ax.spines['bottom'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)

    prev_limit = 0
    for idx2, lim in enumerate(limits):
        ax.barh([1], lim - prev_limit, left=prev_limit, height=h,
                color=palette[idx2])
        prev_limit = lim
    rects = ax.patches
    ax.barh([1], item[1], height=(h / 3), color=bar_color)
    ymin, ymax = ax.get_ylim()
    ax.vlines(item[2], ymin * .9, ymax * .9, linewidth=1.5, color=target_color)

if labels is not None:
    for rect, label in zip(rects, labels):
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2,height * .4,label,ha='center',va='bottom',color=label_color)
if formatter:
    ax.xaxis.set_major_formatter(formatter)
if axis_label:
    ax.set_xlabel(axis_label)
if title:
    fig.suptitle(title, fontsize=14)
fig.subplots_adjust(hspace=0)

plt.savefig("Bullet Chart Python.png")
plt.show()
```
