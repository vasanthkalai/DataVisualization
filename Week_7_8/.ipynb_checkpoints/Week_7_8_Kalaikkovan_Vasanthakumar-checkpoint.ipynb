```python
#importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as py
```

```python
birthdf=pd.read_csv("birth-rate.csv")
crimedf=pd.read_csv("crimerates-by-state-2005.csv")
birthyeardf=pd.read_csv("birth-rates-yearly.csv")
lifedf=pd.read_csv("life-expectancy.csv")
```

### 1. Scatter Plot

```python
plt.figure(figsize=(15,8))
ax=sns.scatterplot(data=crimedf,x="state",y="motor_vehicle_theft").set(title="Motor Vehicle theft vs State")
locs,labels=plt.xticks()
plt.setp(labels,rotation=90)
plt.show()
```

### 2. Bubble plot

```python
plt.figure(figsize=(15,8))
ax=sns.scatterplot(data=crimedf,x="state",y="motor_vehicle_theft", size="motor_vehicle_theft",legend=False,sizes=(20,2000)).set(title="Motor Vehicle theft vs State")
locs,labels=plt.xticks()
plt.setp(labels,rotation=90)
plt.show()
```

### 3. Density Plot

```python
plt.figure(figsize=(15,8))
sns.displot(x=crimedf["motor_vehicle_theft"],kde=True).set(title="Density Plot")
locs,labels=plt.xticks()
plt.setp(labels,rotation=90)
plt.show()
```
